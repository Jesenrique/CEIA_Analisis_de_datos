```python
import pandas as pd

# Creamos un DataFrame con datos de ventas
datos_ventas = {'vendedor': ['Ana', 'Ana', 'Pedro', 'Pedro', 'Juan', 'Juan', 'Ana'],
                'producto': ['A', 'B', 'A', 'B', 'A', 'B', 'B'],
                'cantidad': [3, 4, 1, 2, 5, 3, 2],
                'precio': [100, 50, 80, 120, 70, 60, 55]}
df_ventas = pd.DataFrame(datos_ventas)

# Agrupamos los datos por vendedor y sumamos la cantidad de productos vendidos y los ingresos
agrupado = df_ventas.groupby('vendedor').agg({'cantidad': 'sum', 'precio': 'sum'})

# Imprimimos el resultado
print(agrupado)
```
